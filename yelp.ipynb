evolucion(si las impresiones mejoran con el tiempo)

```python
def tabla_resultado(base_de_datos):
  res=pd.DataFrame()
  res['empresa']=[]
  res['valoracion']=[]
  res['lo_mejor']=[]
  res['lo_peor']=[]
  res['n_comentarios']=[]
  res['lo_mejor_words']=[]
  res['lo_peor_words']=[]
  res.info()

  res['empresa']=df['business_id'].unique() #Generamos la lista de empresas

#Aca sacamos la valoracion promedio que tiene cada empresa
  valoracion=[]
  for i in res['empresa']:
    df_mask=df['business_id']==i
    media=df[df_mask]['stars'].mean()
    media=round(media,1)
    valoracion.append(media)
  res['valoracion']=valoracion

#Es un dato siempre util el saber que cantidad de comentarios avalan los resultados.
  n_comentarios=[]
  for i in res['empresa']:
    df_mask=df['business_id']==i
    n_comentarios.append(df[df_mask].shape[0])
  res['n_comentarios']=n_comentarios

#En esta etapa analizamos el texto. primero lo separamos en oraciones, luego lo clasificamos segun su positividad.
  lo_mejor=[]
  lo_peor=[]
  for i in res['empresa']:
    df_mask=df['business_id']==i
    mejor=[]
    peor=[]
    for opinion in df[df_mask]['text']:
      doc =nlp(opinion)
      for sent in doc.sents:
          if TextBlob(str(sent)).sentiment.polarity > 0.1:#aplico     0.1 como criterio divisor de que es un buen comentario y -0.1 para uno malo
            mejor.append(sent)
          elif TextBlob(str(sent)).sentiment.polarity < -0.1:
            peor.append(sent)
    lo_mejor.append(mejor)
    lo_peor.append(peor)
  res['lo_mejor']=lo_mejor
  res['lo_peor']=lo_peor





  return res
```

```python
resultado=tabla_resultado(df) #Tarda 16 minutos
```
